```python
import pandas as pd
import json
import requests
import numpy as np

data = requests.get("https://raw.githubusercontent.com/ingridcristh/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json")
data = data.json()
```

#ðŸ”§ TransformaciÃ³n

```python
df = pd.json_normalize(data)
df.columns = [col.split('.')[-1] for col in df.columns]
df.head()
```

```python
df.info()
df['Total'] = pd.to_numeric(df['Total'], errors='coerce')

cols = ['PhoneService','MultipleLines']
df[cols] = df[cols].applymap(lambda x: 1 if x == 'Yes' else 0).astype(np.int64)
df.head()
```

```python
df_evasion = df[['customerID','Churn','gender','SeniorCitizen','tenure','PhoneService','MultipleLines','Contract','Monthly','Total']]
df_evasion.dropna()
```
